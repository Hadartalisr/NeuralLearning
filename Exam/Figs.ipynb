```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function for the constraint boundary and its gradient
def constraint(x):
    return np.sqrt(1 - x**2)

def grad_constraint(x):
    return -x / np.sqrt(1 - x**2)

# Define the gradient of the objective function at x*
x_star = np.array([0.6, 0.8])  # Optimal point on the boundary
grad_f_at_x_star = np.array([-1, 0])  # Just an example gradient of f at x*

# Generate points for the constraint boundary
x = np.linspace(-1, 1, 400)
y = constraint(x)

# Start a new figure
plt.figure(figsize=(6, 6))

# Plot the constraint boundary
plt.plot(x, y, label=r'Constraint $g(x) = 0$', color='black')

# Plot the gradient of f at x*
plt.quiver(x_star[0], x_star[1], grad_f_at_x_star[0], grad_f_at_x_star[1], color='red', scale=1, scale_units='xy', width=0.005, label=r'$\nabla f(x^*)$')

# Plot the gradient of g at x*
grad_g_at_x_star = grad_constraint(x_star[0])
plt.quiver(x_star[0], x_star[1], grad_g_at_x_star, 0, color='blue', scale=1, scale_units='xy', width=0.005, label=r'$\nabla g(x^*)$')

# Highlight the optimal point x*
plt.scatter(*x_star, color='green', zorder=5)
plt.text(x_star[0], x_star[1], ' $x^*$', fontsize=12, va='bottom', ha='right')

# Set the aspect ratio of the plot to be equal
plt.axis('equal')

# Set the legend
plt.legend()

# Set the plot limits
plt.xlim(-1.2, 1.2)
plt.ylim(-0.1, 1.2)

# Add grid, title, and labels
plt.grid(True)
plt.title('Illustration of the KKT Condition for a Single Constraint')
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```
